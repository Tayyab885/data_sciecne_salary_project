```python
## Importing Basic Liabraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('glassdoor_jobs.csv',index_col=0)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Salary Estimate'].value_counts()
```

```python
# Let's drop the salary values with -1
df = df[df['Salary Estimate']!='-1']
```

```python
df['Salary Estimate'].value_counts()
```

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0].strip())
salary
```

```python
remove_salary_k = salary.apply(lambda x: x.replace('K','').replace('$',''))
remove_salary_k
```

```python
remove_salary_k.unique()
```

We can see there are some per hourly salary values too as well as Employer Provided Salary.<br>
Let's create a hourly column with 1 if salary estimate have per hour in it else 0.

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df
```

```python
df[df['hourly']==1]
```

Similary with Employer Provided Salary:

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
df
```

```python
df[df['employer_provided']==1]
```

```python
remove_salary_k.unique()
```

```python
## Replace from remove_salary_k too
remove_salary_k_hr = remove_salary_k.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
remove_salary_k_hr.unique()
```

As these salaries are range estimate so let's create min and max salary column.

```python
df['min_salary'] = remove_salary_k_hr.apply(lambda x: x.split('-')[0])
```

```python
df['max_salary'] = remove_salary_k_hr.apply(lambda x: x.split('-')[1])
```

```python
df
```

```python
df['min_salary'] = df['min_salary'].astype(int)
df['max_salary'] = df['max_salary'].astype(int)
```

```python
df.info()
```

```python
df['avg_salary'] = (df['min_salary']+df['max_salary'])/2
df.head()
```

Average Salary will be our target/dependent feature.<br>
So now let's fix company name column

```python
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
```

```python
df.head()
```

```python
## From Location let's just get the State(Last two characters after ,)
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1].strip())
```

```python
df.head()
```

```python
## Check if the company's Headquarters and job is in same place
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

```python
df
```

```python
## Let's get the company age (how long since it's founded)
df['company_age'] = df['Founded'].apply(lambda x: x if x < 1 else 2022 - x)
```

```python
df.head()
```

Parsing Job Description column to get info about required skills(like python, rstudio, spark, aws,excel etc)

```python
df['python_skill'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
df['python_skill'].value_counts()
```

```python
df['spark_skill'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
```

```python
df['aws_skill'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
df['excel_skill'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df.head()
```
